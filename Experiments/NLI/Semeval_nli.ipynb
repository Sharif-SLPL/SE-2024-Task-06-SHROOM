```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

model_name = "MoritzLaurer/DeBERTa-v3-large-mnli-fever-anli-ling-wanli"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name).to(device)

premise = "I first thought that I liked the movie, but upon second thought it was actually disappointing."
hypothesis = "The movie was not good."

input = tokenizer(premise, hypothesis, truncation=True, return_tensors="pt")
output = model(input["input_ids"].to(device))  # device = "cuda:0" or "cpu"
prediction = torch.softmax(output["logits"][0], -1).tolist()
label_names = ["entailment", "neutral", "contradiction"]
prediction = {name: round(float(pred) * 100, 1) for pred, name in zip(prediction, label_names)}
print(prediction)
```

```python
models_name = [
    "MoritzLaurer/DeBERTa-v3-large-mnli-fever-anli-ling-wanli",
    "MoritzLaurer/mDeBERTa-v3-base-mnli-xnli",
    "MoritzLaurer/mDeBERTa-v3-base-xnli-multilingual-nli-2mil7"
]

data_frames = {
    "val_agnostic": val_agnostic,
    "val_aware": val_aware,
    "test": test
}
```

```python
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
import scipy.stats


def AUC_metric(scores, labels, plot=True):
    y = np.where(labels >= 0.5, 1, 0)
    pred = scores
    fpr, tpr, thresholds = metrics.roc_curve(y, pred)
    roc_auc = metrics.auc(fpr, tpr)
    display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,
                                    estimator_name='example estimator')

    if plot:
        display.plot()
        plt.show()
    return roc_auc

def accuracy(scores, labels, threshold=0.5):
    y = np.where(labels >= 0.5, 1, 0)
    pred = np.where(scores >= threshold, 1, 0)

    return metrics.accuracy_score(y, pred)

def S_correlation(scores, labels):
    spearman_corr, p_value = scipy.stats.spearmanr(scores, labels)
    return spearman_corr

def evaluate(name, scores, labels, plot=True):
    print(name, "accuracy:", accuracy(scores, labels))
    print(name, "spearman correlation:", S_correlation(scores, labels))
```

```python
for model_name in models_name:
  tokenizer = AutoTokenizer.from_pretrained(model_name)
  model = AutoModelForSequenceClassification.from_pretrained(model_name).to(device)

  for data_name, data in data_frames.items():
    predictions = {'entailment': [], 'contradiction': []}

    for row in tqdm(data.iterrows()):
      if row[1].task == "PG":
        premise = row[1].src
      elif row[1].task == "MT" or row[1].task == "DM":
        premise = row[1].tgt

      hypothesis = row[1].hyp

      input = tokenizer(premise, hypothesis, truncation=True, return_tensors="pt")
      output = model(input["input_ids"].to(device))
      prediction = torch.softmax(output["logits"][0], -1).tolist()
      label_names = ["entailment", "neutral", "contradiction"]

      predictions['entailment'].append(1 - prediction[0])
      predictions['contradiction'].append(prediction[-1])

    print("******************************")
    print("Model: ", model_name)
    print("Data: ", data_name)
    evaluate('entailment', np.array(predictions['entailment']), np.array(data['p(Hallucination)']))
    print("******")
    evaluate('contradiction', np.array(predictions['contradiction']), np.array(data['p(Hallucination)']))
    print("******************************")
```

```python
datasets = {
    'test.model-agnostic.json': test_agnostic,
    'test.model-aware.json': test_aware
}
```

```python
model_name = "MoritzLaurer/DeBERTa-v3-large-mnli-fever-anli-ling-wanli"
entailment = True

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name).to(device)

for data_name, data in datasets.items():
  output = []
  for row in tqdm(data.iterrows()):
    if row[1].task == "PG":
      premise = row[1].src
    elif row[1].task == "MT" or row[1].task == "DM":
      premise = row[1].tgt

    hypothesis = row[1].hyp

    input = tokenizer(premise, hypothesis, truncation=True, return_tensors="pt")
    model_output = model(input["input_ids"].to(device))
    prediction = torch.softmax(model_output["logits"][0], -1).tolist()
    label_names = ["entailment", "neutral", "contradiction"]

    if entailment:
      prediction = 1 - prediction[0]
    else:
      prediction = prediction[-1]

    output.append({
      'id': row[1].id,
      'label': 'Hallucination' if prediction >= 0.5 else 'Not Hallucination',
      'p(Hallucination)': prediction
    })

  output_path = data_name
  f1 = open(output_path, 'w', encoding='utf-8')
  json.dump(output, f1)
  f1.close()
```
